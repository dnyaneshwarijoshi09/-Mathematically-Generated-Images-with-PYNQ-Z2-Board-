```python
# Install ColabTurtlePlus
!pip install ColabTurtlePlus

from ColabTurtlePlus.Turtle import *

# Initialize screen
initializeTurtle()

# Koch snowflake function
def koch_snowflake(t, length, level):
    if level == 0:
        forward(length)
    else:
        length /= 3.0
        koch_snowflake(t, length, level - 1)
        left(60)
        koch_snowflake(t, length, level - 1)
        right(120)
        koch_snowflake(t, length, level - 1)
        left(60)
        koch_snowflake(t, length, level - 1)

# Set speed and position
speed(0)
penup()
goto(-150, 100)
pendown()

# Draw 3 sides
for i in range(3):
    koch_snowflake(None, 300, 4)
    right(120)

# Keep display
done()
```
